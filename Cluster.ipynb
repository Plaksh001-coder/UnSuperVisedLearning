```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X,y_true=make_blobs(n_samples=500,centers=3,cluster_std=4,random_state=42)
```

```python
df=pd.DataFrame(X,columns=['feature1','feature2'])
```

```python
df
```

```python
sns.scatterplot(data=df,x='feature1',y='feature2')
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
X_scaled
#
```

```python
WCSS=[]
k_range=range(1,11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    WCSS.append(kmeans.inertia_)
```

```python
WCSS
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(k_range, WCSS, marker='o')
```

```python
kmean_final=KMeans(n_clusters=3,random_state=42)
```

```python
cluster_lable=kmean_final.fit_predict(X_scaled)
```

```python
cluster_lable
```

```python
df['cluster']=cluster_lable
```

```python
sns.scatterplot(x=df['feature1'],y=df['feature2'],hue=df['cluster'],palette='viridis')
```

```python
from sklearn.datasets import make_moons
```

```python
X,y_true=make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
from sklearn.cluster import KMeans,DBSCAN
```

```python
df=pd.DataFrame(X,columns=['feature1','feature2'])
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(df)
```

```python
WCSS2=[]
kmeans2_range=range(1,11)
```

```python
for k in kmeans2_range:
  kmeans2=KMeans(n_clusters=k,random_state=42)
  kmeans2.fit(X_scaled)
  WCSS2.append(kmeans2.inertia_)
```

```python
WCSS2
```

```python
plt.plot(kmeans2_range,WCSS2,marker='o')
```

```python
kmeans2_final=KMeans(n_clusters=2,random_state=42)
```

```python
cluster_lable2=kmeans2_final.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=df['feature1'],y=df['feature2'],hue=cluster_lable2,palette='viridis')
```

# DBSCAN

```python
dbscan=DBSCAN(eps=0.3,min_samples=5)
dbscan_lable=dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=df['feature1'],y=df['feature2'],hue=dbscan_lable,palette='viridis')
```
